3. Who scores the most on average for math, reading and writing based on
 Gender
 Test preparation course

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import files
data=files.upload()
```

```python
data
```

```python
df=pd.read_csv('StudentsPerformance.csv')
df
```

```python
plt.figure(figsize=(20,10))
sns.barplot(df['gender'],df['math score'])
plt.xticks(rotation=20)
plt.axhline(y = 66.067, color = 'r', linestyle = 'dashed', label = "avg math score")
plt.axhline(y = 69.117, color = 'g', linestyle = '-', label = "avg reading score")
plt.axhline(y = 67.997, color = 'b', linestyle = ':', label = "avg writing score")
plt.legend(bbox_to_anchor = (1.3, 1), loc = 'upper center')
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.barplot(df['test preparation course'],df['math score'])
plt.xticks(rotation=20)
plt.axhline(y = 66.067, color = 'r', linestyle = 'dashed', label = "avg math score")
plt.axhline(y = 69.117, color = 'g', linestyle = '-', label = "avg reading score")
plt.axhline(y = 67.997, color = 'b', linestyle = ':', label = "avg writing score")
plt.legend(bbox_to_anchor = (1.3, 1), loc = 'upper center')
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.barplot(df['gender'],df['math score'])
plt.xticks(rotation=20)
plt.axhline(y = 66.067, color = 'r', linestyle = 'dashed', label = "var math score")
plt.axhline(y = 69.117, color = 'g', linestyle = '-', label = "var reading score")
plt.axhline(y = 67.997, color = 'b', linestyle = ':', label = "var writing score")
plt.legend(bbox_to_anchor = (1.3, 1), loc = 'upper center')
plt.show()
```
